```python
# 1. Pastikan kolom 'Tgl Dok' bertipe datetime (Langkah kritis)
records_df["Tgl Dok"] = pd.to_datetime(records_df["Tgl Dok"])

# 2. Filter data untuk barang spesifik dan atur 'Tgl Dok' sebagai index
#    Ini menciptakan DataFrame baru dengan index yang benar
df_jack_banana = records_df[
    records_df["Nama Barang"] == "Jack Banana Hitam, merah"
].set_index("Tgl Dok")

# 3. Lakukan Resampling Bulanan ('ME' = Month End) dan Hitung Total 'Jumlah'
#    Sekarang, Anda memanggil resample dari DataFrame yang sudah berindex datetime.
df_bulanan = df_jack_banana["Jumlah"].resample("ME").sum().fillna(0)

print("--- Data Transaksi per Bulan (Resampled) ---")
print(df_bulanan)
```
